```python
with open("snake.txt", "r") as f:
    text = f.read()

text
```

```python
blocks = text.split("\n\n")
blocks = [b.splitlines() for b in blocks]
blocks
```

```python
class Bb:
    def __init__(self, lines: list[str]) -> None:
        if lines[0].startswith("Disassembly"):
            self.title = lines[0]
            self.lines = lines[1:]
        else:
            (index, *rest) = lines[0].split()
            self.title = index
            self.lines = [" ".join(rest)] + lines[1:]

    def __repr__(self) -> str:
        lines = '\n'.join(self.lines)
        return f"TITLE: {self.title}\n-----------\n{lines}\n"

blocks = [Bb(b) for b in blocks]

blocks
```
